Question 1: Do flights delays happen more often later in the day compared to earlier in the day?

To solve this problem, I had to split the 24 hours of the day into subcategories to represent different periods of the day and create a new column: Early morning, Late morning, Early afternoon, Late Afternoon, Early Evening, Late Evening, and Night. The new column called 'TIME_OF_DAY' is added to the original flights dataset. After creating this new column, we can perform some analysis on the column and quickly get some useful insights that tell us that most flights take-off in the morning and afternoon. Moreover, we can also get the normalized values, which tells us that around ~80% of all flights depart in the morning or afternoon, meaning that these hours probably will have more delays. However, just assuming this is not good enough for a data scientist, so I called a function to find total time delayed by the time of the day, and the output showed me that my intuition was right pretty good. Whereas, most flights are delayed in the afternoon or early evening. During the late afternoon is when the most flights are delayed, followed by early evening and late afternoon. This information gives us some potentially useful insights for future references in case I'm looking to travel by air. The best time of the day to fly if you want a smooth and efficient experience seems to be in the morning time considering they have a lot of flights taking off, but not that many are late. However, if you want to try to make sure you are not late, you should probably schedule Night and Late evening flights, considering they have the lowest amount of delayed flights. 

Flights are the least delayed at night and late evening, according to my analysis. However, this can be misrepresentative, considering that there are way fewer flights taking off during the later periods of the day. However, the most frequent delays still appear to occur after noon, considering that the top 3 time periods with the most amount of minutes delayed are: Late Afternoon, Early Evening, Early Afternoon. So yes, the majority of delayed flights will happen in the afternoon. 

```python
# 1. Do flights delays happen more often later in the day compared to earlier in the day?

# Function that separates the day into 7 different categories and creates a new colum: 
# Early morning
# Late morning
# Early afternoon
# Late Afternoon
# Early Evening
# Late Evening
# Night

def get_time_of_day(SCHEDULED_DEPARTURE):
    if SCHEDULED_DEPARTURE < 900 and SCHEDULED_DEPARTURE >= 500:
        return 'EARLY MORNING'
    elif SCHEDULED_DEPARTURE < 1200 and SCHEDULED_DEPARTURE >= 900:
        return 'LATE MORNING'
    elif SCHEDULED_DEPARTURE < 1500 and SCHEDULED_DEPARTURE >= 1200:
        return 'EARLY AFTERNOON'
    elif SCHEDULED_DEPARTURE < 1800 and SCHEDULED_DEPARTURE >= 1500:
        return 'LATE AFTERNOON'
    elif SCHEDULED_DEPARTURE < 2100 and SCHEDULED_DEPARTURE >= 1800:
        return 'EARLY EVENING'
    elif SCHEDULED_DEPARTURE < 2300 and SCHEDULED_DEPARTURE >= 2000:
        return 'LATE EVENING'
    else:
        return 'NIGHT'

flights['TIME_OF_DAY'] = flights['SCHEDULED_DEPARTURE'].apply(get_time_of_day)
```

```python
time_of_day_count = flights['TIME_OF_DAY'].value_counts() # understanding what time of the day most flights depart
time_of_day_count
```

```python
time_of_day_perc = flights['TIME_OF_DAY'].value_counts(normalize=True) * 100 # getting the percentages
time_of_day_perc
```

```python
time_of_day_group = flights.groupby(['TIME_OF_DAY'])
time_of_day_group
```

```python
# finding time delayed by time of day

finding_time_delayed_by_time_of_day = time_of_day_group[['DEPARTURE_DELAY']].sum().sort_values('DEPARTURE_DELAY')
finding_time_delayed_by_time_of_day
```

```python
# Visualization

figure, axes = plt.subplots(figsize=(10,7))

my_colors = 'rgbkymc'

finding_time_delayed_by_time_of_day.plot(ax=axes, kind='bar', stacked=True, color=my_colors) 

axes.get_legend().remove()
axes.set_ylabel('Total Time Delayed in Minutes', fontsize=20)
axes.set_xlabel('Time of the Day', fontsize=20)
axes.set_title('Total Delay time during certain Time Periods', fontsize=25, fontweight='bold')
axes.grid(alpha=.1)
```

Question 2: Does the response to claim 1 depend on the month of the year? 

My response to claim 1, about how flights are more delayed in the afternoon does, in fact, seem to depend on the month of the year. To reach this conclusion, I had to create a visualization representing how delays are different in the AM versus in the PM. As we can see from this visual representation, the most amount of delays occur in frequent air traveling periods during the summer month from May to August, and also during the holiday season in December. The line chart shows how the total amount of time-delayed differs by AM and PM, and by the month of the year. It is interesting to see that the total time delayed for PM is higher than AM for every month, which is a good enough indicator to back my answer that most flights are delayed in the afternoon. 

```python
# Question 2:
# Does the response to claim 1 depend on the month of the year?

# Split day in half to create AM and PM 

def split_day(SCHEDULED_DEPARTURE): 
    if SCHEDULED_DEPARTURE <= 1200 and SCHEDULED_DEPARTURE >=0:
        return 'AM'
    else:
        return 'PM'
flights['AM/PM'] = flights['SCHEDULED_DEPARTURE'].apply(split_day)
```

```python
# Comparing AM and PM total flights

AM_PM_distribution = flights['AM/PM'].value_counts()
AM_PM_distribution
```

```python
# Getting the percentages: 

AM_PM_perc_distribution = flights['AM/PM'].value_counts(normalize=True) * 100
AM_PM_perc_distribution
```

```python
AM_PM_by_month = flights.groupby(['AM/PM', 'MONTH'])

# Summary statistics output to better understand the difference between AM and PM flights
AM_PM_by_month[['DEPARTURE_DELAY']].describe()

# Getting the total delay by each month by AM and PM 
delay_AM_PM_by_month = AM_PM_by_month[['DEPARTURE_DELAY']].sum()
delay_AM_PM_by_month
```

```python
# Visualization

delay_AM = delay_AM_PM_by_month.loc['AM']
delay_PM = delay_AM_PM_by_month.loc['PM']

figure, axes = plt.subplots(figsize=(10,7))

axes.plot(delay_AM.index, delay_AM['DEPARTURE_DELAY'], color = 'green', label='AM', marker='*', alpha=.5)
axes.plot(delay_PM.index, delay_PM['DEPARTURE_DELAY'], color='darkblue', label='PM', marker='o', alpha=.5)

axes.set_xticks(range(1,13))
x_lab = ('Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct'
         , 'Nov', 'Dec')
axes.set_xticklabels(x_lab, rotation=45)

axes.legend()
axes.set_ylabel('Total amount of time in Minutes', fontsize=20)
axes.set_xlabel('Months', fontsize=20)
axes.set_title('Departure Delay by Months', fontsize=25, fontweight='bold')
axes.grid(alpha=.1)
```

This is a visual representation showing how often airlines are delayed with regard to their total amount of trips. A previous graph that I created in this project report represented the total delayed time in the AM/PM and also by airlines. One thing that stood out to me right away was that Southwest had the most mins delayed by far, which made me a little confused because they are supposed to be known for their efficiency. The results showed me, as I suspected, that Southwest is not the worst airline by delayed minutes when, in reality, the US airline is the worst. It is always important not to get fooled by data as it can be misleading.
